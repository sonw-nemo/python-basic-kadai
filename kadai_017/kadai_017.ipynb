```python
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):
  #ageが20以上の場合に大人であること
    if self.age >= 20:
      return "大人"
  #そうでない場合に大人でないこと
    else:
      return "子供"

person = Human("侍太郎",25)

print(person.name)        # 侍太郎
print(person.check_adult())  # True
```
