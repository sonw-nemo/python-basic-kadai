```python
import requests
import pprint
from getpass import getpass

# APIキーを指定
api_key = getpass('APIキーを入力してください: ')
```

```python
# 検索キーワードを変数search_wordに格納
search_word = input('検索キーワードを入力してください: ')
```

```python
# API接続先
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'address':search_word,
        'key': api_key,
    }
)
# レスポンスのJSONを取得
json_data = response.json()

# JSONの中身を確認
pprint.pprint(json_data)

location = json_data['results'][0]['geometry']['location']
lat = location['lat']
lng = location['lng']



```

```python
# レストラン
place_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# データを取得
place_params = {
        'location': f"{lat},{lng}",
        'radius': '500',      # 半径をメートル単位で指定（例：500m）
        'type': 'restaurant', # レストランのみ
        'key': api_key,
        'language': 'ja',
    }
place_response = requests.get(place_url, params=place_params)
place_data = place_response.json()

pprint.pprint(place_data)

# 上で取得したJSONからレストラン情報を3つの項目ごとに表示
for result in place_data['results']:
    name = result.get('name', '情報なし')           # レストラン名
    rating = result.get('rating', '評価なし')       # 評価（星の数など）
    address = result.get('vicinity', '住所なし')    # 住所

    print(f'レストラン名: {name}')
    print(f'評価: {rating}')
    print(f'住所: {address}')
    print('---')

#print(f'{item["name"]}{item["plus_code"]["compound_code"]}')
```
