```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 正の整数を入力してもらう
b = int(input("正の整数を入力してください"))

# 変数varにランダムな正の整数を代入する
var = random.randint(1, b)

# 変数varの値を出力
print(var)

# 変数varが3の倍数かつ5の倍数の場合はFizzBuzzと出力
if var % 3 ==0 and var % 5 ==0 :
  print("FizzBuzz")

# 変数varの値が3の倍数であればFizzと出力
elif var % 3 ==0 :
  print("Fizz")

# 変数varの値が5の倍数であればBuzzと出力
elif var % 5 ==0 :
  print("Buzz")

else :
  print(var)
```
