```python
!pip install openpyxl
import openpyxl

from openpyxl import Workbook
from datetime import date

# ==============================
# Excelファイル作成
# ==============================
wb = Workbook()
ws = wb.active
ws.title = "請求書"

# 日付（YYYY/MM/DD）
today = date(2026, 1, 15).strftime("%Y/%m/%d")

# ==============================
# ヘッダー情報
# ==============================
ws["B2"] = "請求書"

ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL: 03-1234-5678　FAX: 03-1234-5678"
ws["B7"] = "担当者名：鈴木一郎"

ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today

# ==============================
# 明細ヘッダー
# ==============================
headers = ["商品名", "数量", "単価", "金額"]
start_row = 10

for col, header in enumerate(headers, start=2):
    ws.cell(row=start_row, column=col, value=header)

# ==============================
# 商品データ
# ==============================
items = [
    ("商品A", 2, 10000),
    ("商品B", 1, 15000),
]

# 明細行（for + enumerate 使用）
for i, (name, qty, price) in enumerate(items, start=1):
    row = start_row + i
    ws.cell(row=row, column=2, value=name)
    ws.cell(row=row, column=3, value=qty)
    ws.cell(row=row, column=4, value=price)
    ws.cell(row=row, column=5, value=qty * price)

# ==============================
# 合計計算
# ==============================
subtotal = sum(qty * price for _, qty, price in items)
tax = int(subtotal * 0.1)
total = subtotal + tax

total_row = start_row + len(items) + 2

ws[f"D{total_row}"] = "合計"
ws[f"E{total_row}"] = subtotal

ws[f"D{total_row + 1}"] = "消費税"
ws[f"E{total_row + 1}"] = tax

ws[f"D{total_row + 2}"] = "税込合計"
ws[f"E{total_row + 2}"] = total

# ==============================
# 保存
# ==============================
wb.save("請求書_20260115.xlsx")

```
